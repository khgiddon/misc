```python
# Solver for NYT "three-ring" game on Sunday Dec 17 2023 puzzle supplement

path = ''
words = open(path + "words.txt").readlines()
words = [i.strip() for i in words if len(i.strip()) >= 6]

d = {
    0: 'LIT'
    ,1: 'EAI'
    ,2: 'NDA'
    ,3: 'DCT'
    ,4: 'EOI'
    ,5: 'MBA'
    ,6: 'POE'
    ,7: 'OEL'
 }

for word in words:
    for starting_k in d:
        k = starting_k
        for i in range(len(word)):
            if word[i] in d[k].lower():
                k += 1
                if k == 8:
                    k = 0

                if i == len(word) - 1:
                    print(word)
            else:
                break

```
