```python
import random

# https://fivethirtyeight.com/features/can-you-catch-the-grasshopper/

iters = 10000000
x = .5 # Starting x doesn't matter over long time period
landings = dict()
for i in range(0,101):
    landings[i/100] = 0

for _ in range(iters):
    if x >= .2 and x <= .8:
        x += random.uniform(-.2,.2)
    elif x > .8:
        x += random.uniform(-.2,1-x)
    elif x < .2:
        x += random.uniform(-x,.2)
    landings[round(x,2)] += 1

# We've discretized the distribution, so the 0.0 and 1.0 only get half the values they should have gotten under a true continuous process, because of rounding
# So let's multiply them by 2

landings[0.0] *= 2
landings[1.0] *= 2

# Find the ratio
print(max(landings.values())/min(landings.values()))

# 1.9981270067784516
```
