```python
from scipy.stats import hypergeom
import matplotlib.pyplot as plt

# Get results
results = []
maxn = 60
for n in range(22,maxn+2,2):
    [a,b,c] = [n, n/2, 19] # x cards, x/2 red balls, 19 trials
    results.append((n,hypergeom(a,b,c).pmf(8))) # 8 red balls selected
    
maxtuple = max(results,key=lambda item:item[1])
```

```python
# Plot #1
plt.axvline(x=maxtuple[0],linestyle='--', linewidth = 3, color='red')
plt.plot(*zip(*results),linestyle='--', marker='o', color='b')
plt.title('Likelihood of observed result by number of balls')
plt.xlabel('Number of balls')
plt.ylabel('Likelihood')
plt.grid(linestyle='-', linewidth=.5)
plt.rcParams['figure.figsize'] = [10, 7]
plt.xticks(range(22,maxn+2,2))
plt.ylim(bottom=0)
plt.text(35, .13, f'Maximum likelihood at {maxtuple[0]} balls', horizontalalignment='left',verticalalignment='center',fontsize='large',fontstyle='italic')
plt.show()
```
