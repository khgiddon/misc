```python
import random
import matplotlib.pyplot as plt
import math

d = {1:[1]}

def collatz(n):
    
    """

    Function takes in an integer input (>= 1) and returns the Collatz sequence for that integer.
    Stores results (memoizes) in the dictionary d to speed up computation.

    """
    seq = []
    while n not in d:
        seq.append(n)        
        if n % 2 == 0:
            n = n//2
        else:
            n = (3*n)+1
    seq.extend(d[n])
    d[seq[0]] = seq

"""

Generate the first N collatz numbers and then pull the list of sequences

"""

n = 2001
for k in range(1,n):
    collatz(k)
seq_list = list(d.values())


"""

Plot sequence

"""

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, polar=True)
ax.grid(False)
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_facecolor('black')

for v in seq_list:
    theta = [math.radians(t) for t in v]    
    r = [1 for t in v]
    ax.plot(theta,r,color='white',alpha=0.5,linewidth=.015)
plt.show()
```
