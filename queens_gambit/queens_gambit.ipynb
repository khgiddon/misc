```python
import numpy as np
import pandas as pd
import copy
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.style as style
import matplotlib.lines as mlines
import matplotlib.dates as mdates
import matplotlib.ticker as mtick
import datetime as dt


# Graphic params
%matplotlib inline
style.use('fivethirtyeight')
matplotlib.rcParams['figure.dpi'] = 250
matplotlib.rcParams['font.sans-serif'] = ['Arial']

"""

DATA

Data pulled from lichess manually via HTML code on 2023-02-06
by changing settings on Queen's Gambit opening page
https://lichess.org/opening/Queens_Gambit/d4_d5_c4


"""

df = pd.read_csv('data/queens_gambit_lichess_data.csv')
df['month'] = pd.to_datetime(df['month'])
```

```python
print(df.dtypes)
```

```python
# Master controls
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

# Formatting
ax.tick_params(axis = 'both', which = 'major', labelsize = 8)
ax.tick_params(axis='x', rotation=0)
ax.axes.grid(linewidth=0.5)
ax.axvline(x=dt.datetime(2020, 10, 1), color='red', linewidth=1, alpha=.5, linestyle = '--')


x1 = df['month']
y1 = df['all_600_2500']
plt.plot(x1,y1/100,color='blue',alpha=1,linewidth=1.5,linestyle='-')
max_y = max(y1)/100

# Limits
ax.set_ylim([0, max_y*1.1])


# Title
ax.text(x=dt.datetime(2016,5,1),y = max_y*1.1, s = "Mondale's winding route to victory, 1984",
               fontsize = 14, weight = 'bold', alpha = .75)
#ax.text(x=1,y = max_y*1.08, s = """Mondale's victory path with the lowest number of flipped voters includes some surprising states""",
               #fontsize = 8, style = 'italic', alpha = .75)

# Axis labels
ax.set_ylabel("""State's\nelectoral\nvotes""",rotation=0,size=8)
ax.yaxis.set_label_coords(x=-.03,y=1.01)
#ax.set_xlabel("""% of votes in state won by Mondale""",rotation=0,size=8)
#ax.spines['left'].set_visible(False)


ax.yaxis.set_major_formatter(mtick.PercentFormatter(1,decimals=0))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))      
mdates.DateFormatter('%d')
    
# Footer
#ax.text(x = 0.2, y = max_y*-0.22,s = """© Kyle Giddon, www.kylegiddon.com""" + ''.join(' ' for _ in range(27)) + """Sources: MIT Election Data and Science Lab; U.S. National Archive; author's computations""",fontsize = 6, color = '#f0f0f0', backgroundcolor = 'grey')
```
